## 条件分岐

今週は簡単にですが条件分岐という概念を学びます。どのようなプログラムを書くにしろ条件分岐がないものは、基本的に役に立ちません。  
そのくらい重要です。とはいえ、使い方はとても簡単です。

````python
number1 = 15
if number1 > 10:
    print("10より大きい")
if number1 > 20:
    print("20より大きい")

print("終了")
````

## 条件分岐(判定)に使う主な演算子

| 演算子 | 説明 |
| --- | --- |
| `+` | 加算 |
| `-` | 減算 |
| `*` | 乗算 |
| `/` | 除算（浮動小数点数の結果） |
| `//` | 整数除算（切り捨て） |
| `%` | モジュロ（剰余） |
| `**` | べき乗 |
| `=` | 代入 |
| `==` | 等しい |
| `!=` | 等しくない |
| `<` | より小さい |
| `>` | より大きい |
| `<=` | 以下 |
| `>=` | 以上 |
| `and` | 論理和 |
| `or` | 論理積 |
| `not` | 論理否定 |

````python
a = 7
if a == 7 : 
    print('7に等しい')
if a > 4 : 
   print('4より大きい')
if  a >= 7 :
    print('7以上である')
if a < 8 : print('8より小さい') 
if a <= 7 :
    print('7以下である')
if a != 7 :
    print('7と等しくない')

print("終了")
````

## 分岐

条件を判定する時に1つだけではなく、複数の条件で試したいときがあります。そういう時に、`else`という文と`elif`というものを使います。



````python
a = 10
if a > 2 :
    print('2より大きい')
else :
    # a <= 2と同じ
    print('2以下である')

print("終了")
````

````python
a = 10
if a < 2 :
    print('2より小さい')
elif a < 20:
    print('2以上だが、20より小さい')
else :
    print('20以上である')

print("終了")
````

## インデント

英語で文章を書くときは、字下げ(indentation)というものが非常に重要です。余談ですが、字下げをうまく使えるかどうかは海外とのビジネス文書や論文で重要です。  
日本語で段落を変えるときも１文字空けたりしますが、同じ目的です。
何故重要かというと、ここから一つの意味のある固まり(Paragraph)が始まりますよ、と目からも入ってくるので慣れると飛ばし読みしたりする時に重宝します。  
プログラミングでも一つの固まりを作ることが多くあり、コードブロックと呼ばれます。  
どういう時に使うかですが、上のように条件判定など、ひと固まりにしたいときに出てきます。他のほとんどのプログラミング言語では波括弧 `{}`を使いますが、pythonは字下げでブロックを決める珍しい言語です。  

1. **ブロックの開始と終了**：Pythonでは、他の多くのプログラミング言語とは異なり、コードブロック（if文、forループ、関数定義など）の開始と終了を示すために波括弧 `{}` を使用しません。代わりに、インデントがコードブロックの開始と終了を示します。
2. **コードの可読性**：適切なインデントは、コードの構造を明確にし、他の人々（または未来のあなた自身）がコードを理解しやすくします。
3. **構文エラーの回避**：Pythonはインデントを厳密に扱います。インデントが不適切な場合、Pythonは構文エラーを発生させ、プログラムは実行されません。

````python
a = 2
if a < 2 :
    print('2より小さい')
elif a < 20:
    print('2以上だが、20より小さい')
else :
    print('20以上である')

print("終了")
````

## もう少し複雑な条件分岐

なぜインデントが重要なんでしょうか？プログラミングでは入れ子(nested)という状態がよく必要となるからです。  
ifのブロックの中に別のifが出てくること、などです。  
ロシアのマトリョーシカ人形みたいなイメージです。  
余談ですが、ゲノムの解析をしていると生命が入れ子構造を利用しているようなケースをまれに見ます。

````python
a = 10
if a % 2 == 0:
    print('aは偶数である')
    if a > 5:
        print("aは偶数であり、かつ、5より大きい")
else :
    # a % 2 != 0と同じ
    print('2は奇数である')
    if a > 3:
        print("aは奇数であり、かつ、3より大きい")

print("終了")
````

````python
# 入れ子の辞書
# １つの名前で複数の情報を入れたいことがとても良くあり、こういう時に使ったりします。

# 糖度と1kgあたりの価格(円)
data = {"banana":
          {"sugar-content": 0.2, "price": 332},
        "apple":
          {"sugar-content": 0.15, "price": 741}
       }

#入れ子辞書を使って入れ子の判定

if data['banana']['price'] > data['apple']['price']:
    print("バナナのほうが1kgあたりの値段が高い")
    if data['banana']['sugar-content'] > data['apple']['sugar-content']:
        print("バナナは高いが、より甘い")
    else:
        print("バナナは高くて、さらに甘さが少ない!")
elif data['banana']['price'] < data['apple']['price']:
    print("リンゴのほうが1kgあたりの値段が高い")
    if data['banana']['sugar-content'] < data['apple']['sugar-content']:
        print("リンゴは高いが、より甘い")
    else:
        print("リンゴは高くて、さらに甘さが少ない!")
else:
    print("バナナとリンゴは１kgあたり同じ値段である")
    if data['banana']['sugar-content'] == data['apple']['sugar-content']:
        print("バナナとリンゴは同じ値段で、甘さも同じ")
    else:
        print("バナナとリンゴは同じ値段だが、甘さは異なる")

print("終了")
````
